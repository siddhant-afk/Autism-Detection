##### Adding Categorical Fields

0 - Adults
1 - Adolescent
2 - Children

```python
df1["Age Group"] = 0
df1.head()
```

```python
df2["Age Group"] = 1
df2.head()
```

```python
df3["Age Group"] = 2
df3.head()
```

```python
merged_df = pd.merge(df1,df2,how="outer")
merged_df = pd.merge(merged_df,df3,how="outer")
# merged_df.reset_index(inplace=True)
merged_df.head()
```

```python
len(merged_df)
```

```python
merged_df["Age Group"].value_counts()
```

### Exploratory Data Analysis

```python
data = merged_df
```

```python
print("No. of records : ", len(data.index))
print("No. of individuals diagnosed with Autism : ", len(data[data["Class/ASD"] == b"YES"]))
print("No. of individuals not diagnosed with Autism : ",len(data[data["Class/ASD"] == b"NO"]))
print(f'Percentage of individuals diagnosed with Autism :  {len(data[data["Class/ASD"] == b"YES"])/len(data.index) * 100:.2f}%')
```

```python
print("No. of Null fields:\n",data.isna().sum())
```

```python
data.info()
```

```python
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
sns.violinplot(x="jundice", y="result", hue="austim", data=data, split=True,
                inner="quart", palette={b'yes': "r", b'no': "b"})
sns.despine(left=True)
```

```python
sns.violinplot(x="jundice", y="result", hue="Class/ASD", data=data, split=True,
                inner="quart", palette={b'YES': "r", b'NO': "b"})
sns.despine(left=True)
```

```python
sns.violinplot(x="gender", y="result", hue="Class/ASD", data=data, split=True,
                inner="quart", palette={b'YES': "r", b'NO': "b"})
sns.despine(left=True)
```
